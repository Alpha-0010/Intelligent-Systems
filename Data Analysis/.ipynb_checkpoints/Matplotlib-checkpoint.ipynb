```python
# Matplolib pyplot is a collection of command style functions that make matplotlib work like MATLAB.
import matplotlib.pyplot as plt
```

```python
# Example
plt.plot([2, 4, 6, 4]) # plot(y_list)
plt.ylabel("Numbers")
plt.xlabel("Indices")
plt.title("My plot")
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16]) # plot(x_list, y_list)
plt.xlabel("Squares")
plt.ylabel("Numbers")
plt.grid() # grid on
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # plot(x_list, y_list, red and dots)
plt.grid()
plt.show()
```

```python
import numpy as np
t = np.arange(0., 5., 0.2)

# blue dashed, red squares and green triangles
plt.plot(t, t ** 2, 'b--', label='^2')
plt.plot(t, t ** 2.2, 'rs', label='^2.2')
plt.plot(t, t ** 2.5, 'g^', label='^2.5')
plt.grid()
plt.legend() # add legend based on line labels -- Legend expains which line represents what
plt.show()
```

```python
# Controlling line width
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth = 5.0)
plt.show()
```

```python
# Line properties -- Check the documentation to check more properties

x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]

lines = plt.plot(x1, y1, x2, y2)
# Using keyword args like Python
plt.setp(lines[0], color = 'r', linewidth = 2.0)
# Using String value pairs like MATLAB, we can use either of the technoques.
plt.setp(lines[1], 'color', 'g', 'linewidth', 2.0)
plt.grid()
plt.show()
```

```python
# Breaking a single figure into 2 parts
import numpy as np
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi*t)

t1 = np.arange(0.0, 5.0, 1.0)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)

# First subplot
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'r--')


# Second subplot
plt.subplot(212)
plt.plot(t2, np.cos(2 * np.pi * t2), 'b-')

plt.show()
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot([1, 2, 3])
plt.subplot(212)
plt.plot([4, 5, 6])

plt.figure(2)
plt.plot([4, 5, 6])

# Go back to figure - 1 subplot - 1 and do some operations
plt.figure(1)
plt.subplot(211)
plt.title('Easy as 1, 2, 3')


plt.show()
```
