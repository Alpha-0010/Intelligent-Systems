```python
# Concatenating Data Frames
import pandas as pd
india_weather = pd.DataFrame(
    {
        "city" : ["Mumbai", "Delhi", "Bangalore"],
        "temperature" : [32, 45, 30],
        "windspeed" : [80, 60, 78]
    }
)

us_weather = pd.DataFrame(
    {
        "city" : ["New York", "Chicago", "Orlando"],
        "temperature" : [21, 14, 35],
        "windspeed" : [68, 65, 75]
    }
)

df = pd.concat([india_weather, us_weather])
df
```

```python
# If we want continous indices in the concatenated DF
df = pd.concat([india_weather, us_weather], ignore_index = True)
df
```

```python
# Concatenate DFs row wise
df = pd.concat([india_weather, us_weather], axis = 1)
df
```

```python
# Merge DataFrames --> It is like Join Command in SQL
temperature_df = pd.DataFrame(
    {
        "city" : ["Mumbai", "Delhi", "Bangalore", "Hyderabad"],
        "temperature" : [32, 45, 30, 40]
    }
)
humidity_df = pd.DataFrame(
    {
        "city" : ["Delhi", "Mumbai", "Bangalore"],
        "humidity" : [68, 65, 76]
    }
)
# Merge 2 DFs without explicitly mention index (INNER_JOIN)
df = pd.merge(temperature_df, humidity_df, on = 'city')
print(df)
# OUTER_JOIN
print("************************************************************")
df2 = pd.merge(temperature_df, humidity_df, on = 'city', how = 'outer')
print(df2)
```

```python
# Numerical Indexing (.loc vs iloc)
# If we don't give the index parameter then the indexing will be 0 based.
df = pd.DataFrame([1, 2, 3, 4, 5], index = [10, 20, 30, 40, 50])
print(df)
print("*****************************")
# loc -> Use index to get the value
print(df.loc[40])
print("*****************************")
# iloc -> Get the data present in this row
print(df.iloc[3])
```

```python
# Drop duplicates from the DataFrame which have the same email-id and keep only the first occurence
df = pd.DataFrame({
    'Name' : ['Alex', 'Bob', 'Alice' , 'Jack', None],
    'email-id' : ['alex@gmail.com', 'bob@gmail.com', 'alice@gmail.com', 'alex@gmail.com', 'na@gmail.com']
})

# keep can be first, last or False --> False will mean dropping all the duplicates rows
df.drop_duplicates(subset = ['email-id'], keep = 'first', inplace = True)
print(df)

print("**************************************************************")
# Drop rows with NA (NONE) name in the df
df_cleaned = df.dropna(subset = ['Name'])
print(df_cleaned)

# Rename columns
df.rename(columns = {'Name' : 'Full Name'}, inplace = True)
df
```

```python
import pandas as pd
df = pd.DataFrame(
    {
        'Name' : ['Alex', 'Bob', 'Alice', 'Carlize'],
        'Grade' : [90.2, 88.7, 45, 99.8]
    }
)

# Change data-type of a column
df['Grade'] = df['Grade'].astype('int')
print(df)

employees = pd.DataFrame(
    {
        'id' : [1, 2, 3],
        'Score': [None, 90, 70]
    }
)

print("*************************************************")

# Fill the None values with 0 in the marks columns
employees['Score'] = employees['Score'].fillna(0)
print(employees)
```

```python
# Pivot a table in Pandas

import pandas as pd
weather = pd.DataFrame({
    'city' : ['Mumbai', 'Nyc', 'Dublin', 'Mumbai', 'Dublin', 'Dublin', 'Nyc', 'Nyc', 'Mumbai', 'Mumbai'],
    'temperature' : [40, 30, 28, 30, 18, 10, 18, 9, 19, 11],
    'month' : ['Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'June', 'Mar', 'June', 'June', 'Oct'] 
})

print(weather)
print("******************************************")
# Pivot the DF weather such that each row will represent a specific month, each column will represent the city and we will have the temperature for that
# speific city and month in that cell.

pivot_weather = pd.pivot_table(weather, index = 'month', columns = 'city', values = 'temperature', fill_value = 0)
pivot_weather
```
