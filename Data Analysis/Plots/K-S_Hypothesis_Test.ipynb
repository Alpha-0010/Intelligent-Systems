```python
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

#generate a gaussian r.v X
x = stats.norm.rvs(size=1000);
print(len(x),type(x))
sns.set_style('whitegrid')
sns.kdeplot(x, bw=0.5)
plt.show()
```

```python
stats.kstest(x, 'norm')

# Since, the p-value is very high 'X' follows Normal Distribution.
```

```python
# Y ~ Continous Uniform Distribution(0,1)
y = np.random.uniform(0,1,10000);
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')

# P-value is 0, so y doesn't follow Normal Distribution.
```
